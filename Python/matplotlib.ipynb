```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
arr = np.random.randint(20000,40000,100)
arr
```

```python
#lineplot --- numerical data
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker = "s",ms = 10,
         mfc = "red",mec = "green",
         color = "black", lw = 3, ls = ":",
         alpha = 0.8, label =  "lineplot")
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#scatterplot --- numerical data based 
plt.scatter(x,y, marker = "s")
plt.show()
```

```python
#histogram -- distribution n of data in intervals(bins)
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python
##piechart --- percentage
x = ["mp","up","rj","mh"]
y = [10,20,30,40]
explodes = (0.2,0,0,0)
plt.pie(y,labels = x,autopct = "%1.2f%%",shadow=True,explode= explodes)
plt.savefig("piechart.pdf")
plt.show()
#savefig:

#plt.savefig("piechart.pdf")
```

```python
#subplot
x = np.linspace(0,10,100)
plt.figure(figsize = (10,6))
plt.subplot(2,2,1) ## 2 rows or 2 columns: 1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)  ## 2 rows or 2 columns: 2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)  ## 2 rows or 2 columns: 3rd graph
plt.plot(x,np.tan(x))  
plt.title("tan")
plt.subplot(2,2,4)  ## 2 rows or 2 columns: 4th graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()

```
